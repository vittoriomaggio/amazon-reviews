# Riduzione dataset #

```python
import numpy as np
import pandas as pd
import string
import itertools
import matplotlib.pyplot as plt

products = pd.read_csv("products_clean.csv")
reviews = pd.read_csv("reviews_clean.csv")

setCategories = ["music", "musical-instruments", "books", "english-books", "dvd", "videogames"]
```

Voglio trovare il numero di prodotti per categoria

```python
numProducts = pd.DataFrame()

numProducts["Categoria"] = setCategories

listCount = []
for cat in setCategories:
    count = 0
    for pcat in products.category:
          if (cat == pcat):
            count = count + 1
    listCount.append(count)
    
numProducts["num_prodotti"] = listCount    
```

```python
numProducts
```

```python
prodDataset = pd.DataFrame()

prodDataset["Categoria"] = list(products.category)
prodDataset["product"] = list(products._id)
```

```python
result = pd.merge(prodDataset, reviews)
```

```python
listCount2 = []
for cat in setCategories:
    count2 = 0
    for rcat in result.Categoria:
          if (cat == rcat):
            count2 = count2 + 1
    listCount2.append(count2)
    
numProducts["num_recensioni"] = listCount2    
```
