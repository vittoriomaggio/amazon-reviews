# List of products with most number of positive and negative Sentiment (3 Different list). (3 per ogni lista) #

```python
pos_Dataset = pd.DataFrame()
neg_Dataset = pd.DataFrame()

pos_Dataset['title'] = dataset['title']
pos_Dataset['sentiment'] = dataset['sentiment']

neg_Dataset['title'] = dataset['title']
neg_Dataset['sentiment'] = dataset['sentiment']

pos_Dataset = pos_Dataset.loc[dataset['sentiment'] == "positive"]
neg_Dataset = neg_Dataset.loc[dataset['sentiment'] == "negative"]

pos_result = pos_Dataset.groupby('title').count()
neg_result = neg_Dataset.groupby('title').count()
```

```python
pos_list = pos_result['sentiment'].sort_values(ascending=False)
pos_list[0:3]
```

```python
fig_pos = go.Figure(go.Bar(
            x=[pos_list[0], pos_list[1], pos_list[2]],
            y=['The Lord of the Rings - The Motion Picture Trilogy, Extended Edition   ', 
               'Grand Theft Auto V (GTA V) - PlayStation 4   ',
               'FIFA 19 - PlayStation 4   '],
            orientation='h'))

fig_pos.show()
```

```python
neg_list = neg_result['sentiment'].sort_values(ascending=False)
neg_list[0:3]
```

```python
fig_neg = go.Figure(go.Bar(
            x=[neg_list[0], neg_list[1], neg_list[2]],
            y=['Gran Turismo Sport - PlayStation 4   ', 
               'Vivere 120 anni. Le verità che nessuno vuole raccontarti   ',
               'La verità sul caso Harry Quebert   '],
            orientation='h'))

fig_neg.show()
```
